```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.tree import DecisionTreeClassifier,export_graphviz
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn.ensemble import RandomForestClassifier
```

```python
kanser=pd.read_csv("kanser.csv")
```

```python
kanser.head(3)
```

```python
kanser.info()
```

```python
kanser=kanser.drop("id",axis=1)
```

```python
kanser.head(3)
```

```python
kanser["diagnosis"]=np.where(kanser["diagnosis"]=="M",0,1)
kanser.head(3)
```

```python
y=kanser[["diagnosis"]]
x=kanser.drop("diagnosis",axis=1)
```

```python
x_train,x_test,y_train, y_test= train_test_split(x,y,test_size=0.8,random_state=5)
```

```python
rf=RandomForestClassifier()
model=rf.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
tree=DecisionTreeClassifier()
model=tree.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
tree=DecisionTreeClassifier()
model=tree.fit(x,y)
model.score(x,y)
```
